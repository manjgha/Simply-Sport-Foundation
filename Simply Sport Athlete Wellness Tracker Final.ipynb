```python
#Converting symptoms column into an array with all its values
symptoms_array= np.array(athlete_df['Symptoms']).flatten()
symptoms_array
```

```python
#Extracting unique symptoms from symptoms array
import re

def extract_symptoms(symptom_string):
    matches = re.findall(r"'([^']*)'", symptom_string)
    return matches

all_symptoms = []
for item in symptoms_array:
    symptoms = extract_symptoms(item)
    all_symptoms.extend(symptoms)

unique_symptoms = sorted(set(all_symptoms))

print(f"Found {len(unique_symptoms)} unique symptoms:")
print(unique_symptoms)
```

```python
#Making each symptom into a new column and filling values into each column (1 if the athlete experienced the symptom, 0 otherwise)

unique_symptoms = ['Anxiety', 'Back pain', 'Bloating', 'Body pain', 'Cramps', 'Crying', 
                  'Dizziness', 'Fatigue', 'Hard to focus', 'Headache', 'Irritability', 
                  'Loose motion', 'Mood swings', 'Nausea', 'No hunger', 'Overeating', 
                  'Stomach pain']

def check_symptom(symptom_list, target_symptom):
    # For string representation of lists
    if isinstance(symptom_list, str):
        return 1 if f"'{target_symptom}'" in symptom_list else 0
    else:
        return 0


for symptom in unique_symptoms:
    athlete_df[symptom] = athlete_df['Symptoms'].apply(
        lambda x: check_symptom(x, symptom)
    )
```

```python
athlete_df.head()
```

```python
#Creating another table, solely for symptoms
symptoms_df= athlete_df[['Date', 'Athlete', 'Symptoms']]
symptoms_df.head()
```

```python
import ast

symptoms_df['Symptoms'] = symptoms_df['Symptoms'].apply(ast.literal_eval)
elongated_df = symptoms_df.explode('Symptoms')
elongated_df = elongated_df.reset_index(drop=True)
elongated_df.head()
```

```python
elongated_df.shape, elongated_df.info()
```

```python
elongated_df['Symptoms']=elongated_df['Symptoms'].fillna('No symptoms')
elongated_df.info()
```

```python
#Getting unique values and no. of unique values for each column

cols=elongated_df.columns

for i in cols:
    print(f'The number of unique values in {i} are: {elongated_df[i].nunique()}. \n')
    print(f'The unique values in {i} are: {elongated_df[i].unique()}. \n\n')
```

```python
# Extracting year, month, and day into new columns for symptoms table
elongated_df['Month'] = elongated_df['Date'].dt.month
elongated_df['Day'] = elongated_df['Date'].dt.day

elongated_df['Month_Name'] = elongated_df['Date'].dt.month_name()

elongated_df.head()
```

```python
# Extracting year, month, and day into new columns for athletes table

athlete_df['Month'] = athlete_df['Date'].dt.month
athlete_df['Day'] = athlete_df['Date'].dt.day

athlete_df['Month_Name'] = athlete_df['Date'].dt.month_name()
athlete_df.head()
```

```python
#Writing each table into 1 excel file
athlete_df.to_excel('Athlete Wellness Preprocessed.xlsx', sheet_name='Athlete Data')

file_path = "D:\OneDrive\Documents\Course\Imarticus Learning\Simply Sport\Athlete Wellness Preprocessed.xlsx"

# Writing elongated_df into a new sheet in the previously created excel file
with pd.ExcelWriter(file_path, engine='openpyxl', mode='a') as writer:
    elongated_df.to_excel(writer, sheet_name='Symptoms', index=False)

print("Data exported successfully!")
```
